```pyspark
from pyspark.sql import functions as F

# 1. Yıldız dağılımı
print("=== Yıldız Dağılımı ===")
df.groupBy("star_rating").count().orderBy("star_rating").show()

# 2. Helpful vote oranı istatistikleri
print("=== Helpful/Total Vote İstatistikleri ===")
df.select(
    F.mean("helpful_votes").alias("avg_helpful"),
    F.mean("total_votes").alias("avg_total"),
    F.mean(F.when(F.col("total_votes") > 0,
        F.col("helpful_votes") / F.col("total_votes")
    )).alias("avg_helpful_ratio")
).show()

# 3. Review body uzunluğu
print("=== Review Uzunluğu İstatistikleri ===")
df.withColumn("review_len", F.length("review_body")).select(
    F.mean("review_len").alias("avg_len"),
    F.min("review_len").alias("min_len"),
    F.max("review_len").alias("max_len"),
    F.percentile_approx("review_len", 0.5).alias("median_len")
).show()

# 4. Aynı müşterinin kaç farklı ürüne yorum yaptığı (spam sinyali)
print("=== Müşteri Başına Yorum Sayısı (Top 10) ===")
df.groupBy("customer_id").count().orderBy(F.desc("count")).show(10)
```

```pyspark
from pyspark.sql import functions as F

# Fake review label stratejisi
# Kriter 1: Extreme yıldız (1 veya 5)
# Kriter 2: Düşük helpful ratio (oy almış ama yararlı bulunmamış)
# Kriter 3: Çok kısa review body

df_labeled = df.withColumn(
    "helpful_ratio",
    F.when(F.col("total_votes") > 0,
        F.col("helpful_votes") / F.col("total_votes")
    ).otherwise(F.lit(None))
).withColumn(
    "review_len",
    F.length("review_body")
).withColumn(
    "is_fake",
    F.when(
        (F.col("star_rating").isin(1, 5)) &
        (F.col("review_len") < 20) &
        (
            (F.col("helpful_ratio") < 0.3) |
            F.col("helpful_ratio").isNull()
        ),
        1
    ).otherwise(0)
)

# Label dağılımı
total = df_labeled.count()
fake_count = df_labeled.filter(F.col("is_fake") == 1).count()
real_count = total - fake_count

print("=== Label Dağılımı ===")
print(f"Toplam  : {total:,}")
print(f"Fake (1): {fake_count:,}  ({100*fake_count/total:.1f}%)")
print(f"Real (0): {real_count:,}  ({100*real_count/total:.1f}%)")

df_labeled.groupBy("is_fake", "star_rating").count().orderBy("is_fake", "star_rating").show(20)
```

```pyspark
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
import time

# Sadece ihtiyacımız olan sütunları al
df_model = df_labeled.select("review_body", "is_fake") \
    .withColumnRenamed("is_fake", "label")

# Train/Test split (%80/%20)
train_df, test_df = df_model.randomSplit([0.8, 0.2], seed=42)
print(f"Train: {train_df.count():,} | Test: {test_df.count():,}")

# Pipeline: Tokenize → StopWords → TF → IDF → LR
tokenizer = Tokenizer(inputCol="review_body", outputCol="words")
remover = StopWordsRemover(inputCol="words", outputCol="filtered_words")
hashingTF = HashingTF(inputCol="filtered_words", outputCol="tf_features", numFeatures=10000)
idf = IDF(inputCol="tf_features", outputCol="features")
lr = LogisticRegression(maxIter=10, regParam=0.01)

pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, idf, lr])

# Eğitim süresi ölç (throughput için)
start_time = time.time()
model = pipeline.fit(train_df)
train_time = time.time() - start_time
print(f"Eğitim süresi: {train_time:.1f} saniye")

# Tahmin
predictions = model.transform(test_df)

# Metrikler
binary_eval = BinaryClassificationEvaluator(labelCol="label")
multi_eval = MulticlassClassificationEvaluator(labelCol="label")

auc = binary_eval.evaluate(predictions)
accuracy = multi_eval.evaluate(predictions, {multi_eval.metricName: "accuracy"})
f1 = multi_eval.evaluate(predictions, {multi_eval.metricName: "f1"})
precision = multi_eval.evaluate(predictions, {multi_eval.metricName: "weightedPrecision"})
recall = multi_eval.evaluate(predictions, {multi_eval.metricName: "weightedRecall"})

# Throughput
test_count = test_df.count()
throughput = test_count / train_time

print("\n=== Logistic Regression Sonuçları ===")
print(f"Accuracy  : {accuracy:.4f}")
print(f"F1-Score  : {f1:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"AUC-ROC   : {auc:.4f}")
print(f"\nThroughput: {throughput:,.0f} satır/saniye")
```

```pyspark
from pyspark.ml.classification import RandomForestClassifier
import time

# Aynı pipeline'ı RF ile kur (TF-IDF features hazır)
rf = RandomForestClassifier(
    numTrees=50,
    maxDepth=10,
    labelCol="label",
    seed=42
)

pipeline_rf = Pipeline(stages=[tokenizer, remover, hashingTF, idf, rf])

start_time = time.time()
model_rf = pipeline_rf.fit(train_df)
train_time_rf = time.time() - start_time
print(f"RF Eğitim süresi: {train_time_rf:.1f} saniye")

predictions_rf = model_rf.transform(test_df)

auc_rf = binary_eval.evaluate(predictions_rf)
accuracy_rf = multi_eval.evaluate(predictions_rf, {multi_eval.metricName: "accuracy"})
f1_rf = multi_eval.evaluate(predictions_rf, {multi_eval.metricName: "f1"})
precision_rf = multi_eval.evaluate(predictions_rf, {multi_eval.metricName: "weightedPrecision"})
recall_rf = multi_eval.evaluate(predictions_rf, {multi_eval.metricName: "weightedRecall"})

throughput_rf = test_df.count() / train_time_rf

print("\n=== Random Forest Sonuçları ===")
print(f"Accuracy  : {accuracy_rf:.4f}")
print(f"F1-Score  : {f1_rf:.4f}")
print(f"Precision : {precision_rf:.4f}")
print(f"Recall    : {recall_rf:.4f}")
print(f"AUC-ROC   : {auc_rf:.4f}")
print(f"\nThroughput: {throughput_rf:,.0f} satır/saniye")

# Karşılaştırma tablosu
print("\n=== Model Karşılaştırması ===")
print(f"{'Metrik':<12} {'LR':>8} {'RF':>8}")
print("-" * 30)
print(f"{'Accuracy':<12} {accuracy:>8.4f} {accuracy_rf:>8.4f}")
print(f"{'F1-Score':<12} {f1:>8.4f} {f1_rf:>8.4f}")
print(f"{'Precision':<12} {precision:>8.4f} {precision_rf:>8.4f}")
print(f"{'Recall':<12} {recall:>8.4f} {recall_rf:>8.4f}")
print(f"{'AUC-ROC':<12} {auc:>8.4f} {auc_rf:>8.4f}")
print(f"{'Throughput':<12} {10585:>8,} {throughput_rf:>8,.0f}")
```

```pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
import time

spark = SparkSession.builder \
    .config("spark.sql.shuffle.partitions", "200") \
    .config("spark.executor.memoryOverhead", "1g") \
    .getOrCreate()

bucket_name = "fake-review-detection-proje-042026"
df = spark.read.parquet(f"s3://{bucket_name}/processed-data/reviews_final_parquet")

# Label üret
df_labeled = df.withColumn("review_len", F.length("review_body")) \
    .withColumn("helpful_ratio",
        F.when(F.col("total_votes") > 0,
            F.col("helpful_votes") / F.col("total_votes"))
        .otherwise(None)) \
    .withColumn("label",
        F.when(
            F.col("star_rating").isin(1, 5) &
            (F.col("review_len") < 20) &
            (F.col("helpful_ratio").isNull() | (F.col("helpful_ratio") < 0.3)),
            1).otherwise(0))

df_model = df_labeled.select("review_body", "label")
train_df, test_df = df_model.randomSplit([0.8, 0.2], seed=42)

# RF için küçük sample
train_rf = train_df.sample(0.05, seed=42)
test_rf  = test_df.sample(0.05, seed=42)

# Pipeline bileşenleri
tokenizer = Tokenizer(inputCol="review_body", outputCol="words")
remover   = StopWordsRemover(inputCol="words", outputCol="filtered_words")
hashingTF = HashingTF(inputCol="filtered_words", outputCol="tf_features", numFeatures=5000)
idf       = IDF(inputCol="tf_features", outputCol="features")

binary_eval = BinaryClassificationEvaluator(labelCol="label")
multi_eval  = MulticlassClassificationEvaluator(labelCol="label")

def get_metrics(predictions, train_time, n_test, label="Model"):
    print(f"\n=== {label} ===")
    print(f"Accuracy  : {multi_eval.evaluate(predictions, {multi_eval.metricName: 'accuracy'}):.4f}")
    print(f"F1-Score  : {multi_eval.evaluate(predictions, {multi_eval.metricName: 'f1'}):.4f}")
    print(f"Precision : {multi_eval.evaluate(predictions, {multi_eval.metricName: 'weightedPrecision'}):.4f}")
    print(f"Recall    : {multi_eval.evaluate(predictions, {multi_eval.metricName: 'weightedRecall'}):.4f}")
    print(f"AUC-ROC   : {binary_eval.evaluate(predictions):.4f}")
    print(f"Throughput: {n_test/train_time:,.0f} satır/saniye")

# --- Logistic Regression ---
print("LR eğitimi başlıyor...")
lr = LogisticRegression(maxIter=10, regParam=0.01)
pipeline_lr = Pipeline(stages=[tokenizer, remover, hashingTF, idf, lr])
t0 = time.time()
model_lr = pipeline_lr.fit(train_df)
lr_time = time.time() - t0
pred_lr = model_lr.transform(test_df)
n_test_lr = test_df.count()
get_metrics(pred_lr, lr_time, n_test_lr, "Logistic Regression (full data)")

# --- Random Forest ---
print("\nRF eğitimi başlıyor (5% sample)...")
rf = RandomForestClassifier(numTrees=10, maxDepth=6, labelCol="label", seed=42)
pipeline_rf = Pipeline(stages=[tokenizer, remover, hashingTF, idf, rf])
t0 = time.time()
model_rf = pipeline_rf.fit(train_rf)
rf_time = time.time() - t0
pred_rf = model_rf.transform(test_rf)
n_test_rf = test_rf.count()
get_metrics(pred_rf, rf_time, n_test_rf, "Random Forest (5% sample)")

print("\nTamamlandı!")
```

```pyspark
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline
import time

# Stratified sample: fake ve real'den eşit oranda al
fake_sample = train_df.filter(F.col("label") == 1).sample(0.5, seed=42)
real_sample = train_df.filter(F.col("label") == 0).sample(0.03, seed=42)
train_rf_balanced = fake_sample.union(real_sample)

fake_test = test_df.filter(F.col("label") == 1).sample(0.5, seed=42)
real_test = test_df.filter(F.col("label") == 0).sample(0.03, seed=42)
test_rf_balanced = fake_test.union(real_test)

# Dağılımı kontrol et
total = train_rf_balanced.count()
fake_n = train_rf_balanced.filter(F.col("label") == 1).count()
print(f"Balanced train - Fake: {fake_n:,} | Real: {total-fake_n:,} | Oran: {100*fake_n/total:.1f}%")

rf = RandomForestClassifier(
    numTrees=10,
    maxDepth=6,
    maxBins=32,
    labelCol="label",
    seed=42
)
pipeline_rf = Pipeline(stages=[tokenizer, remover, hashingTF, idf, rf])

t0 = time.time()
model_rf = pipeline_rf.fit(train_rf_balanced)
rf_time = time.time() - t0
pred_rf = model_rf.transform(test_rf_balanced)
n_test_rf = test_rf_balanced.count()

get_metrics(pred_rf, rf_time, n_test_rf, "Random Forest (balanced sample)")

# Karşılaştırma tablosu
print("\n=== Final Karşılaştırma ===")
print(f"{'Metrik':<12} {'LR (full)':>12} {'RF (balanced)':>14}")
print("-" * 40)
metrics = ["accuracy", "f1", "weightedPrecision", "weightedRecall"]
names   = ["Accuracy", "F1-Score", "Precision", "Recall"]
for name, m in zip(names, metrics):
    lr_val = multi_eval.evaluate(pred_lr, {multi_eval.metricName: m})
    rf_val = multi_eval.evaluate(pred_rf, {multi_eval.metricName: m})
    print(f"{name:<12} {lr_val:>12.4f} {rf_val:>14.4f}")
lr_auc = binary_eval.evaluate(pred_lr)
rf_auc = binary_eval.evaluate(pred_rf)
print(f"{'AUC-ROC':<12} {lr_auc:>12.4f} {rf_auc:>14.4f}")
print(f"{'Throughput':<12} {12531:>12,} {n_test_rf/rf_time:>14,.0f}")
```
